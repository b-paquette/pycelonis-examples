## Create 3D plots with your IBC data

```python
# import the required libraries

import pandas as pd
from pycelonis import get_celonis
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

#### Connect to celonis and get the analysis component containing the 3 dimensions you want to plot

```python
celonis = get_celonis()

analysis= celonis.analyses.find("0d9ed060-2a44-4a8a-a350-ab54444ff2a8")
df = analysis.draft.components.find('Test plot').get_data_frame()
```

#### Create 3D Scatter plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100


xs = pd.to_numeric(df['BKPF.GJAHR'])
ys = pd.to_numeric(df['BSEG.DMBTR'])
zs = pd.to_numeric(df['Total throughput time in days'])
ax.scatter(xs, ys, zs)
ax.set_xlabel('Year')
ax.set_ylabel('DMBTR')
ax.set_zlabel('Throughput time in days')
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```
